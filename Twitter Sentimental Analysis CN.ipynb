```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot
```

```python
#Importing the Data
train_data = pd.read_csv(r"training_twitter_x_y_train.csv")
test_data = pd.read_csv(r"test_twitter_x_test.csv")
```

```python
import nltk, string
```

```python
xtest=test_data["text"]
xtest=np.array(xtest)
xtest=xtest.reshape(len(xtest),1)
xtest.shape

test_tweets=[]
c=0
for text in xtest:
    c+=1
    test_tweets.append(nltk.word_tokenize(str(text)))
```

```python
test_tweets
```

```python
tweets = []
for i in train_data.index:
    tweets.append( ( nltk.word_tokenize(train_data['text'][i]), train_data['airline_sentiment'][i] ) )
```

```python
print(tweets[0:5])
```

```python
import random
random.shuffle(tweets)
random.shuffle(test_tweets)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag

def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
w = "testing"
pos_tag([w])[0][1]
```

```python
useless_words = nltk.corpus.stopwords.words("english") + list(string.punctuation)
```

```python
def clean_tweets(words):
    output_words = []
    
    for w in words:
        if w.lower() not in useless_words:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    
    return output_words     
```

```python
train_tweets = [(clean_tweets(tweet), category) for tweet, category in tweets]
```

```python
test_tweets = [clean_tweets(tweet) for tweet in test_tweets ]
```

```python
training_tweets = train_tweets
testing_tweets = test_tweets
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
categories = [category for tweet, category in training_tweets]
```

```python
text_tweets = [" ".join(tweet) for tweet, category in training_tweets]
test_text_tweets = [" ".join(tweet) for tweet in testing_tweets]
```

```python
count_vec = CountVectorizer(max_features = 7000)
x_train_features = count_vec.fit_transform(text_tweets)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features = count_vec.transform(test_text_tweets)
x_test_features
```

```python
xtrain = x_train_features
ytrain = categories
xtest = x_test_features
```

```python
from sklearn.svm import SVC
svc = SVC(C=0.2,kernel='linear')
svc.fit(xtrain, ytrain)
```

```python
svc = svc.predict(xtest)

svc_np = np.array(svc)
np.savetxt("predicted_sentiments.csv",svc, fmt='%s',encoding=None)
```
